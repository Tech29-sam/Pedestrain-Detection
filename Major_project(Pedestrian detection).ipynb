```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import sys
```

```python
full_body_cascade = cv2.CascadeClassifier('haarcascade_fullbody.xml')
upper_body_cascade = cv2.CascadeClassifier('haarcascade_upperbody.xml')
#full_body_cascade = cv2.CascadeClassifier('pedestrian.xml')
```

```python
ped_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_fullbody.xml')
```

```python
ped_cascade
```

```python
cap = cv2.VideoCapture("pedestrians.avi")
while True:
    ret, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    body = ped_cascade.detectMultiScale(gray,1.2,5)
    for (x,y,w,h) in body:
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    cv2.imshow('video',img)
    k = cv2.waitKey(10) & 0xff
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
